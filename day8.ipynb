```python
#线性回归练习(一元线性回归)穷举法为例
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#数据集
x_data=[1,2,3]
y_data=[3,6,9]

#随机初始化权重和偏置
w = 0
```

```python
#向前传播
def forward(x):
    return x * w

#计算损失函数
def loss(x,y):
    y_pred=forward(x)
    return (y_pred-y)*(y_pred-y)
```

```python
#训练模型
w_list=[]
mse_list=[]
for w in np.arange(0,6.1,0.1):#设置0-6的范围，步长为0.1
    print("w=",w)
    l_sum=0#损失值求和
    for x_val,y_val in zip(x_data,y_data):
        y_pred_val=forward(x_val)#y预测值
        loss_val=loss(x_val,y_val)#损失值
        l_sum+=loss_val#损失值求和
        print("\t",x_val,y_val,y_pred_val,loss_val)
    print("MSE=",l_sum/3)
    w_list.append(w)
    mse_list.append(l_sum/3)
    
```

```python
#数据可视化
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
