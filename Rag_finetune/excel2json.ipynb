# 糖尿病数据集转换


````python
# 加载excel文件
import pandas as pd

# 加载xls文件
file_path = "C:/Users/Admin/Desktop/Data/糖尿病问题_39健康网.xls"
df = pd.read_excel(file_path, engine="xlrd")

# 显示前几行数据
print(df.head())
````

目标格式

```
[
  {
    "conversations": [
      {
        "from": "human",
        "value": "问题一"
      },
      {
        "from": "gpt",
        "value": "模型回答1"
      }
    ],
    "system": "",
    "tools": " "
  }，

    "conversations": [

      }

    "system": "",

      {

        "from": "human",
        "value": "问题二"
      },
      {
        "from": "gpt",
        "value": "模型回答1"
      }
    ],
    "system": "",
    "tools": ""
  }
]
```


````python
import json

# 每个问题和对应的多个回答分别转换成独立的 JSON 对象
# 假设第四列是human提问，第五到第十二列是gpt回答
all_conversations = []
for index, row in df.iterrows():
    human_question = row.iloc[3]
    # 检查human_question是否为NaN，如果是则替换为空字符串
    if pd.isna(human_question):
        human_question = ""

    # 遍历第五到第十二列的回答
    for i in range(4, 12):
        if pd.notna(row.iloc[i]):  # 检查是否为空数据
            gpt_answer = row.iloc[i]
            conversations = [
                {"from": "human", "value": human_question},
                {"from": "gpt", "value": gpt_answer},
            ]
            data = {"conversations": conversations, "system": "", "tools": ""}
            all_conversations.append(data)


output_file = "C:/Users/Admin/Desktop/Data/diabetes_train.json"
# 保存为json文件
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(all_conversations, f, ensure_ascii=False, indent=4)

print(f"转换完成，已保存为{output_file}")
````
